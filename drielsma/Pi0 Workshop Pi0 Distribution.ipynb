## Energy and angular distribution of $\pi^0$

In the rest frame, the two gammas produced from the $\pi^0$ decay each carry half of the pion mass, $m_{\pi^0}/2$, and are emitted back-to-back isotropically around the pion. With a boost factor $\gamma$, the energy distribution of the pion is uniformly distributed from 0 to $E_{\pi^0}$, i.e.

$$\frac{dN}{dE_\gamma} = \frac{2}{E_\pi^0}$$.

The angle between the two gammas is $\pi$ at rest but may be larger when the $\pi^0$ is boosted:

$$\frac{dN}{d\theta} = \frac{1}{4\beta\gamma}\frac{\cos(\theta/2)}{\sin^2(\theta/2)}\frac{1}{\sqrt{\gamma^2\sin^2(\theta/2) - 1}} = \frac{1}{4\beta\gamma}\frac{\sqrt{\cos\theta}}{1-\cos\theta}\frac{1}{\sqrt{\gamma^2(1-\cos\theta) - 1}} $$,

which peaks at $\cos\theta=1-1/\gamma^2=\beta^2$ (minimum allowed angle) and vanishes at $\pi$.

```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn

seaborn.set(rc={
    'figure.figsize':(15, 10),
})
seaborn.set_context('talk') # or paper

dist = lambda x, g: np.sqrt(x)/(1-x)/np.sqrt(g**2*(1-x)-1)/(4*g*np.sqrt(1-1/g**2))

for g in np.arange(1.5, 6., 1.):
    x = np.arange(0, 1.-1./g**2, 0.01)
    y = dist(x, g)
    plt.plot(x, y)

plt.xlabel('cos(theta)')
plt.ylabel('dN/dcos(theta)')
plt.gca().set_ylim([0,2])
```

```python
x = np.arange(0, 1., 0.05)
y = np.arange(1.0001, 6., .25)
x, y = np.meshgrid(x, y)
z = dist(x, y)

# Plot the surface.
import plotly.graph_objects as go
from plotly.offline import iplot
surface = go.Surface(x=x, y=y, z=z)
fig = go.Figure([surface])
fig.update_layout(scene = dict(
        xaxis = dict(nticks=4, range=[0,1],),
                     yaxis = dict(nticks=4, range=[0,6],),
                     zaxis = dict(nticks=4, range=[0,2],),),
                     width=700,
                     margin=dict(r=20, l=10, b=10, t=10))
iplot(fig)
```
